 ### Description: 
 A large-scale study, known as the GBD Study 2010, assessed the impact of various diseases, injuries, and risk factors on mortality rates around the world. This collaborative effort, led by the Institute for Health Metrics and Evaluation (IHME), involved several prestigious universities and the World Health Organization (WHO).
The study examined death counts and mortality rates in 187 countries, 21 regions, and the entire globe. This data encompassed various age groups and genders for five years: 1970, 1980, 1990, 2000, and 2010. Additionally, the study investigated the likelihood of death between ages 15 and 60 (45q15) and between birth and age 5 (5q0). It also calculated average lifespans for these locations and years, categorized by se

The findings of this comprehensive study were published in a prestigious medical journal, The Lancet, in December 2012, under the title "Age-specific and sex-specific mortality in 187 countries, 1970-2010: a systematic analysis for the GBD Study 2

## Bar graphs of number per deaths and death rate per 100,000
I generated 2 bar graphs for the 2 columns I am working with. My interpretation here is that the number of death rate per 100,000 was the most highest in 1970. 1980 was the highest in terms of number of deaths.

```python

df['Number of Deaths'] = pd.to_numeric(df['Number of Deaths'], errors='coerce')
df['Death Rate Per 100,000'] = pd.to_numeric(df['Death Rate Per 100,000'], errors='coerce')
df = df.dropna(subset=['Number of Deaths', 'Death Rate Per 100,000'])

yearly_data = df.groupby('Year').agg({'Number of Deaths': 'sum', 'Death Rate Per 100,000': 'sum'})

fig, axes = plt.subplots(2, 1, figsize=(10, 12))

# Plot total number of deaths per year
axes[0].bar(yearly_data.index, yearly_data['Number of Deaths'], color='skyblue', edgecolor='black')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Total Number of Deaths')
axes[0].set_title('Bar Graph of the Total Number of Deaths per Year')

# Plot death rate per 100,000 per year
axes[1].bar(yearly_data.index, yearly_data['Death Rate Per 100,000'], color='lightgreen', edgecolor='black')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Death Rate Per 100,000')
axes[1].set_title('Bar Graph of Death Rate Per 100,000 per Year')

plt.tight_layout()
plt.show()
```

### Piecharts

## Piechart for top countries death rate per 100,000
I wanted to compare countries and death rates per 100,000 using a piechart to see precisely the percentages 

```python

# Clean the 'Death Rate Per 100,000' column (if necessary)
if not pd.api.types.is_numeric_dtype(df['Death Rate Per 100,000']):
    df['Death Rate Per 100,000'] = pd.to_numeric(df['Death Rate Per 100,000'].str.replace(',', ''), errors='coerce')  # Handle potential errors

# Sort by Death Rate Per 100,000 (descending order)
df_sorted = df.sort_values(by='Death Rate Per 100,000', ascending=False)

# Select top 10 countries (or all if less than 10)
top_10 = min(10, len(df_sorted))
countries = df_sorted['Country Name'].head(top_10).to_numpy()
death_rates = df_sorted['Death Rate Per 100,000'].head(top_10).to_numpy()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(death_rates, labels=countries, autopct="%1.1f%%", startangle=140)  # Adjust startangle for better readability

# Customize plot elements
plt.title('Top Countries by Death Rate Per 100,000 (2010)')
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart
plt.tight_layout()
plt.show()
```

## Piechart of top countries by number of deaths

I wanted to compare countries and number of deaths using a piechart to see precisely the percentages 

```python

if not pd.api.types.is_numeric_dtype(df['Number of Deaths']):
    df['Number of Deaths'] = pd.to_numeric(df['Number of Deaths'].str.replace(',', ''), errors='coerce')
if not pd.api.types.is_numeric_dtype(df['Death Rate Per 100,000']):
    df['Death Rate Per 100,000'] = pd.to_numeric(df['Death Rate Per 100,000'].str.replace(',', ''), errors='coerce')


death_rate_metric = 'Number of Deaths'  
df_sorted = df.sort_values(by=death_rate_metric, ascending=False)

top_10 = min(10, len(df_sorted))
countries = df_sorted['Country Name'].head(top_10).to_numpy()
death_rates = df_sorted[death_rate_metric].head(top_10).to_numpy()

plt.figure(figsize=(8, 8))
plt.pie(death_rates, labels=countries, autopct="%1.1f%%", startangle=140)  

plt.title(f'Top Countries by {death_rate_metric} (2010)')
plt.axis('equal') 
plt.tight_layout()
plt.show()
```
